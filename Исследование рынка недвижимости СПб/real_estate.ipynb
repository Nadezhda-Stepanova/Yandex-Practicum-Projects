# Исследование объявлений о продаже квартир

В вашем распоряжении данные сервиса Яндекс.Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Ваша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

## Изучение данных из файла

```python
import pandas as pd

import warnings                           # отключили предупреждения        
warnings.filterwarnings('ignore')  


data = pd.read_csv('real_estate_data.csv', sep='\t')

display(data.head(15))      # Печатаем первые 15 строк
data.info()                 # Выводим информацию о таблице
```

В каждой строке таблицы данные о квартире. Таблица имеет 22 столбца, согласно описанию:
-	airports_nearest — расстояние до ближайшего аэропорта в метрах (м)
-	balcony — число балконов
-	ceiling_height — высота потолков (м)
-	cityCenters_nearest — расстояние до центра города (м)
-	days_exposition — сколько дней было размещено объявление (от публикации до снятия)
-	first_day_exposition — дата публикации
-	floor — этаж
-	floors_total — всего этажей в доме
-	is_apartment — апартаменты (булев тип)
-	kitchen_area — площадь кухни в квадратных метрах (м²)
-	last_price — цена на момент снятия с публикации
-	living_area — жилая площадь в квадратных метрах (м²)
-	locality_name — название населённого пункта
-	open_plan — свободная планировка (булев тип)
-	parks_around3000 — число парков в радиусе 3 км
-	parks_nearest — расстояние до ближайшего парка (м)
-	ponds_around3000 — число водоёмов в радиусе 3 км
-	ponds_nearest — расстояние до ближайшего водоёма (м)
-	rooms — число комнат
-	studio — квартира-студия (булев тип)
-	total_area — площадь квартиры в квадратных метрах (м²)
-	total_images — число фотографий квартиры в объявлении

Из них 2 имеют булево значение, а не 3, как заявлено в описании (is_apartment имеет тип object). Большая часть данных записана в формате float.

Из первоначальных данных видно, что имеется несколько проблем.
- Только 8 столбцов из 22 не имеют пропусков.
- Формат даты включает не нужные нулевые значения часы:минуты:секунды, но нет даты недели.
- В названии столбца cityCenters_nearest строчные буквы сочетаются с прописными, а для улучшения читаемости не используется символ "_".
- Значительную часть данных из типа float следует перевести в тип int (например, общее количество этажей или число балконов, кроме площади в м²).
- Не хватает некоторых данных для анализа, например, соотношения жилой и общей площади, варианта этажности ("первый", "последний", "другой").


### Вывод

Предварительно можно утверждать, что данных достаточно для проверки гипотез. Но встречаются пропуски в данных, в названиях колонок есть расхождения с хорошим стилем, формат данных часто не соответствует требуемому представлению, не хватает некоторых высчитанных столбцов с данными.
Сначала нужно устранить эти проблемы.

## Предобработка данных

```python
data = data.rename(columns={'cityCenters_nearest' : 'city_centers_nearest'})  # Переименовали столбец
```

Проверяем, нет ли отрицательных числовых значений. Рассмотрим минимальные значения.
Для этого применим метод min() к необходимым количественным столбцам.

```python
list_columns = ['total_images', 'last_price', 'total_area', # Создали список столбцов с числовыми значениями
                'rooms', 'ceiling_height', 'floors_total', 'living_area', 'floor', 
                'kitchen_area', 'balcony', 'airports_nearest', 'city_centers_nearest', 
                'parks_around3000', 'parks_nearest', 'ponds_around3000', 'ponds_nearest', 'days_exposition'] 

print(data[list_columns].min())   
```

Отрицательных значений нет.

Посмотрим, как представлены значения параметра "аппартаменты".

```python
print(data['is_apartment'].unique())
```

Видим, что принимаемые значения только nan, False, True. Заменим nan на False, так как аппартаменты встречаются очень редко, и логично предположить, когда это аппартаменты, это указывают, а если просто квартира, то значение пропускают. Переведем из object в булево значение, заодно удалив пропуски. Посчитаем количество аппартаментов.

```python
data['is_apartment'] = data['is_apartment'].fillna(False)
print('Всего аппартаментов:', data['is_apartment'].sum())
```

Теперь приведем формат времени столбца first_day_exposition к datetime64. Отбросим значения часов, минут, секунд (округлим до дней). 

```python
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S') 
                                                                                      #  Перевели в формат даты

data['first_day_exposition'] = data['first_day_exposition'].dt.round('1d')            # Округлили до дня, отбросив часы
```

Проверяем, что получилось в итоге.

```python
data.info()
display(data.sample(5))  # вывели пять произвольных строк датафрейма.
```

### О пропусках


Итак. Данные имеют пропуски. Заполним их в таблице следующим образом.

ceiling_height - высота потолков, возьмем медиану, так как из предыдущего вызова функции describe() видно, что есть слишком большое и похожее на вброс максимальное значение, влияющее на среднее.

```python
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median()) # Заменили NaN на медиану
```

floors_total - этажность дома, пропусков 86. В общем объеме на статистику значительно не повлияет, даже если оставить пропуск, а категорию из "первый", "другой", "последний" указать "другой". Но мы заменим NaN на 0 и в дальнейшем при категоризации этажа учтем, чтобы такая квартира попала в категорию "другой". 
Проверять, верно ли указаны данные, и больше ли общее число этажей, чем этаж квартиры, мы не будем, так как даже не верно указанные данные не повлияют на присвоение категории "другой".

```python
data['floors_total'] = data['floors_total'].fillna(0)             # Заменили NaN на 0
```

balcony - количество балконов, считаем, что если не указано, то их нет, то есть заменяем на 0.

```python
data['balcony'] = data['balcony'].fillna(0)                       # Заменили NaN на 0
```

Формально столбец 'rooms' пропусков NaN не имеет, но минимальное значение 0 говорит о том, что данные заполнены не корректно. Квартира имеет хотя бы одну комнату, это мы и укажем. Даже если возможен вариант, что 0 - это студия, у студии тоже есть комната.

```python
data.loc[data['rooms'] == 0, 'rooms'] = 1                         # Заменили 0 на 1
```

living_area, kitchen_area - жилая площадь и площадь кухни. Так как общая площадь известна для всех квартир, можно посчитать соотношение жилая/общая и кухня/общая для известных строк, найти их медиану и применить ее к вычислению площади кухни и жилой для тех строк, где известны только общая площадь или общая и один из параметров.

Новые столбцы 'percent_living_total_area' и 'percent_kitchen_total_area' помогут нам в дальнейших исследованиях.

```python
# Добавили столбец, который пригодится и при дальнейших исследованиях.
data['percent_living_total_area'] = data['living_area'] / data['total_area'] 
     
# Переменная - медианное значение отношения жилой площади к общей:    
med_percent_living_total_area = data['percent_living_total_area'].median()

print('Доля жилой площади к общей составляет примерно', round(med_percent_living_total_area,2))

data['living_area'] = data['living_area'].fillna(data['total_area']*med_percent_living_total_area)
```

```python
# Добавили еще один столбец, который пригодится и далее.
data['percent_kitchen_total_area'] = data['kitchen_area'] / data['total_area'] 
     
med_percent_kitchen_total_area = data['percent_kitchen_total_area'].median() # Аналогично кухня / общая площадь
print('Доля площади кухни к общей составляет примерно', round(med_percent_kitchen_total_area,2))

data['kitchen_area'] = data['kitchen_area'].fillna(data['total_area']*med_percent_kitchen_total_area)
```

Площадь любых видов кухонь не входит в жилую площадь квартиры и при определении «студии» не учитывается. Поэтому для студий получившуюся или уже имевшуюся площадь кухни (явная ошибка) заменим на 0.

Пересчитаем новые столбцы с учетом заполнений.

```python
data.loc[data['studio'] == True, 'kitchen_area'] = 0  # Заменили площадь кухни на 0

data['percent_living_total_area'] = data['living_area'] / data['total_area'] # Пересчитали столбцы, теперь и они без NaN
data['percent_kitchen_total_area'] = data['kitchen_area'] / data['total_area']
```

Ради интереса посчитаем количество квартир, в которых доля кухни и доля жилой площади больше 1, если такие есть.

```python
print(data[(data['percent_living_total_area'] + data['percent_kitchen_total_area']) > 1]['total_area'].count())
```

Так как при замене с коэффициентами 0,57 и 0,17 больше 1 получить невозможно, это ошибка в первоначальной базе данных. Строки ошибочные, их не много, менее 1%, их можно и лучше удалить.

```python
data = data.loc[data['percent_living_total_area'] + data['percent_kitchen_total_area'] <= 1] # Сделали срез по условию
```

locality_name  - населенный пункт. Пропущенных значений всего 44. При объеме данных в 23699 строк это ничтожно малое количество. А так как анализировать данные, не имея информации, город ли это, деревня и т.п, не имеет смысла, эти строки просто удалим.
Обновим индексацию после удаления строк.

```python
data = data.dropna(subset=['locality_name']).reset_index(drop=True) # Применили метод dropna() и обновили индексацию
```

Так как дальнейшие рассуждения будут идти в зависимости от группировки по населенным пунктам, удалим неявные дубликаты в названиях.

### Дубликаты в названиях

```python
print(data['locality_name'].unique()) 
print(len(data['locality_name'].unique())) # Выведем список уникальных названий и их число
```

Видим, что следует удалить слова типа "посёлок", "село" и т.п

```python
for name in [' ', 'посёлок', 'городской', 'деревня', 'поселок', 'городского', 
             'типа', 'садовое', 'товарищество', 'село', 'садоводческое', 'некоммерческое',
            'при', 'железнодорожной', 'станции', 'коттеджный']:
    data['locality_name'] = data['locality_name'].str.replace(name, '')          # Заменили на ""
    
print(len(data['locality_name'].unique()))                 # Проверяем еще раз, сколько уникальных значений осталось.
```

Число уникальных значений снизилось почти на 60.

### Снова пропуски

Теперь возвращаемся к заполнению значений.

airports_nearest и city_сenters_nearest - имеет смысл заменить на медианные значения для данного населенного пункта. Если это небольшой поселок, то даже при одном известном расстоянии до центра или аэропорта медиана даст хорошее приближение именно ввиду компактного размера населенного пункта. 

Создадим таблицу медианных расстояний для населенных пунктов методом groupby().

```python
median_airports_nearest = data.groupby('locality_name')['airports_nearest'].median()  # группируем

print(median_airports_nearest.head()) #  Выведем первые 5 строк
```

Видим, что значений гораздо меньше, чем населенных пунктов. Значит, если медиана одного населенного пункта  NaN, заменим на общую медиану всех населенных пунктов, так как это, в общем-то, одна георгафическая область с единственным аэропортом.

Посмотрим, сколько уникальных значений в столбце расстояния до аэропорта.

```python
print(data['airports_nearest'].isna().sum())  # Пропусков по столбцу
```

```python
for town in data['locality_name'].unique():  # Заполняем пропуски медианами по нас.пунктам (в том числе значениями NaN)
    
    data.loc[(data['airports_nearest'].isna()) & (data['locality_name'] == town), 'airports_nearest'] = median_airports_nearest[town]
    
print(data['airports_nearest'].isna().sum()) # Проверяем, насколько уменьшилось число пропусков.
```

Пропусков стало меньше, теперь остальные заполним общей медианой.

```python
data['airports_nearest'] = data['airports_nearest'].fillna(data['airports_nearest'].median())  
# оставшиеся заменили на общую медиану
print(data['airports_nearest'].isna().sum()) # Проверяем число пропусков.
```

city_сenters_nearest - проверили, система работает, поступаем аналогично.

```python
median_city_сenters_nearest = data.groupby('locality_name')['city_centers_nearest'].median()
for town in data['locality_name'].unique():  # Заполняем пропуски медианами по нас.пунктам (в том числе значениями NaN)
    data.loc[(data['city_centers_nearest'].isna()) & 
             (data['locality_name'] == town), 
             'city_centers_nearest'] = median_city_сenters_nearest[town]
    
data['city_centers_nearest'] = data['city_centers_nearest'].fillna(data['city_centers_nearest'].median())  

print(data['city_centers_nearest'].isna().sum()) # Проверяем оставшееся число пропусков.
```

parks_around3000, ponds_around3000 - количество парков и прудов поблизости; поступим для них, как с балконами, то есть будем считать, что если не указано, то их нет (заменим на 0). Но присвоим значение 1, если вдруг соответствующее значение parks_nearest или ponds_nearest есть и меньше или равно 3000.

```python
data['parks_around3000'] = data['parks_around3000'].fillna(0)
data.loc[(data['parks_around3000'] == 0) & (data['parks_nearest'] <= 3000), 'parks_around3000'] = 1
```

```python
data['ponds_around3000'] = data['ponds_around3000'].fillna(0)
data.loc[(data['ponds_around3000'] == 0) & (data['ponds_nearest'] <= 3000), 'ponds_around3000'] = 1
```

parks_nearest и ponds_nearest - расстояния в метрах до парков \ прудов, пропуски заменим на медиану по нас. пунктам, если количество парков \ прудов рядом больше 0. 
Остальные значения оставим NaN, так как они не повлияют на дальнейшее исследование. Заменять на 0 не корректно, на медиану по области тоже.

```python
median_parks_nearest = data.groupby('locality_name')['parks_nearest'].median()
for town in data['locality_name'].unique():  # Заполняем пропуски медианами по нас.пунктам (в том числе значениями NaN)
    data.loc[(data['parks_nearest'].isna()) & 
             (data['locality_name'] == town) & 
             (data['parks_around3000']>0), 'parks_nearest'] = median_parks_nearest[town]

print(data['parks_nearest'].isna().sum()) # Проверяем оставшееся число пропусков.
```

```python
median_days_exposition = data.groupby('locality_name')['days_exposition'].median()
for town in data['locality_name'].unique():  # Заполняем пропуски медианами по нас.пунктам (в том числе значениями NaN)
    data.loc[(data['days_exposition'].isna()) & (data['locality_name'] == town), 'days_exposition'] = median_days_exposition[town]
    
data['days_exposition'] = data['days_exposition'].fillna(data['days_exposition'].median())  

print(data['days_exposition'].isna().sum()) # Проверяем оставшееся число пропусков.
```

### Сравнение выводов по квартирам в центре и общих выводов по всему городу

Рассмотрим сегмент квартир в центре. Как мы выяснили, это фонд в 8-километровом радиусе от центра.
Проанализируем эту территорию и изучим следующие параметры: площадь, цена, число комнат, высота потолков. 

```python
data_spb_center = data_spb[data_spb['city_centers_nearest_round'] <= 8]  # Выделили значения расстояния до центра <= 8 км
```

Площадь квартир посчитаем, как и в прошлый раз, спользуя квартили и межквартильный размах.

```python
q1 = data_spb_center['total_area'].quantile(0.25) 
q3 = data_spb_center['total_area'].quantile(0.75)
iqr = q3 - q1 # Расчет IQR


# применим присвоение параметру ax первой гистораммы, чтобы вывести старую гистограмму и новую на одном графике:
   
ax = (data.
      hist('total_area', 
           histtype='step', 
           bins=30, 
           figsize=(8, 5), 
           range=(0, q3 + 1.5 * iqr)
          )
     )

(data_spb_center.
 hist('total_area', 
      ax=ax,               # присвоили ax=ax, чтобы вывести на одном графике
      bins=30, 
      figsize=(8, 5), 
      range=(0, q3 + 1.5 * iqr)
     )   
)
```

Видим, что если для исследования в целом был характерен всплеск общей площади в районе 42 кв. м, то для центра города (желтая гистограмма) размер квартир достаточно равномерно распределяется в пределах от 25 до 120 м, то есть выбор квартир любого необходимого метража шире.

Посмотрим, как изменяется стоимость кв. метра.

```python
q1 = data_spb_center['price_m2'].quantile(0.25) 
q3 = data_spb_center['price_m2'].quantile(0.75)
iqr = q3 - q1 # Расчет IQR

# Старая гистограмма:

ax = (data.
      hist('price_m2', 
           bins=30, 
           histtype='step', 
           figsize=(8, 3), 
           range=(q1 - 1.5 * iqr, q3 + 1.5 * iqr))
     )

# Новая:

(data_spb_center.
 hist('price_m2', 
      bins=30, 
      ax=ax, 
      figsize=(8, 3), 
      range=(q1 - 1.5 * iqr, q3 + 1.5 * iqr))
)
```

Для комнат в центре города характерно более равномерное распределение по стоимости, причем цена за метр варьируется в диапазоне примерно от 75 до 170 тысяч за метр, тогда как цена в общем по данным от 35 до 140 тысяч. Центр города ценится, квартиры стоят дороже.
